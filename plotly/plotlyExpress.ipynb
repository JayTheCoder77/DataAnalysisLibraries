```python
import plotly.express as px
import pandas as pd
df = pd.read_csv('/content/iris.csv')
df
```

```python
px.pie(df,names="Species")
```

```python
px.histogram(df,x='SepalWidthCm',nbins=2,title='Histogram of SepalWidthCm')
```

```python
px.box(df,
       y = 'PetalLengthCm',
       title = 'petal len',
       hover_data=['Id']
       )
```

```python
fig = px.scatter(df, x='SepalLengthCm',y='PetalLengthCm')
fig.update_traces(marker=dict(
        size=16,line=dict(width=2,color='DarkSlateGray')
    ), selector=dict(mode='markers'))
fig.show()
```

```python
px.bar(df,x='Species',y='SepalLengthCm',color='PetalWidthCm')
```

```python
import numpy as np
x , y = np.arange(10) , np.arange(10,20)
px.scatter(x=x,y=y)
```

```python
import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(go.Scatter(x=x,y=x,mode='markers',name='Linear'))
fig.add_trace(go.Scatter(x=x,y=y**2,mode='markers',name='Quadratic'))
fig.update_layout(title='Linear vs quadratic')
```

```python
fig = go.Figure()
x = np.arange(-30,30)
fig.add_trace(go.Scatter(x=x,y=x,name='Linear'))
fig.add_trace(go.Scatter(x=x,y=x**2,name='Quadratic'))
fig.update_layout(title='Linear vs quadratic')
```

```python
from plotly.subplots import make_subplots

fig = make_subplots(rows=2,cols=2,subplot_titles=['Linear','Quadratic','Cubic','Quartic'])
fig.add_trace(go.Scatter(x=x,y=x**1,mode="markers"),
              row=1,col=1)
fig.add_trace(go.Scatter(x=x,y=x**2,mode="markers"),
              row=1,col=2)
fig.add_trace(go.Scatter(x=x,y=x**3,mode="markers"),
              row=2,col=1)
fig.add_trace(go.Scatter(x=x,y=x**4,mode="markers"),
              row=2,col=2)
fig.show()
```
