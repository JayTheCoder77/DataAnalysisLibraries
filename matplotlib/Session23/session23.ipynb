# Data **Visualization**
**Types of data**

1.   Numerical Data
2.   Categorical Data



```python
# import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2d plot
# bivariate analysis
# categorical -> numerical and numerical -> numerical
# use case -> time series data
```

```python
# plotting a simple function
price = [48000,54000,57000,49000,47000,45000]
years = [2015,2016,2017,2018,2019,2020]
plt.plot(years , price)
```

```python
# from a pandas datframe
batsman = pd.read_csv('sharma-kohli.csv')
batsman
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# mutiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
# labels title
plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Ipl Seasons')
plt.ylabel('Runs Scored')
```

```python
#  colors (hex) and line ( width and height )
# mutiple plots
plt.plot(batsman['index'],batsman['V Kohli'] ,color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#ef6b1d')
# labels title
plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Ipl Seasons')
plt.ylabel('Runs Scored')
```

```python
# line style
#  colors (hex) and line ( width and height )
# mutiple plots
# plt.plot(batsman['index'],batsman['V Kohli'] ,color='green',linestyle='dashed')
# plt.plot(batsman['index'],batsman['V Kohli'] ,color='green',linestyle='dashdot')
plt.plot(batsman['index'],batsman['V Kohli'] ,color='green',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#ef6b1d',linestyle='dotted',linewidth=2,marker='+')
# labels title
plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Ipl Seasons')
plt.ylabel('Runs Scored')
```

```python
# legend -> location
plt.plot(batsman['index'],batsman['V Kohli'] ,color='green',label='VIRAT')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#ef6b1d',label='ROHIT')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Ipl Seasons')
plt.ylabel('Runs Scored')

plt.legend(loc='upper right')
```

```python
# limiting axes
price = [48000,54000,57000,49000,47000,45000,4500000]
years = [2015,2016,2017,2018,2019,2020,2021]
plt.plot(years , price)
# graph has outlier

plt.ylim(0,175000)
plt.xlim(2017,2019)
```

```python
# grid

plt.plot(batsman['index'],batsman['V Kohli'] ,color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#ef6b1d')
# labels title
plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Ipl Seasons')
plt.ylabel('Runs Scored')

plt.grid()
```

```python
# show
plt.plot(batsman['index'],batsman['V Kohli'] ,color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#ef6b1d')
# labels title
plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Ipl Seasons')
plt.ylabel('Runs Scored')

plt.show()
```
