```python
# import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2d plot
# bivariate analysis
# categorical -> numerical and numerical -> numerical
# use case -> time series data
```

```python
# plotting a simple function
price = [48000,54000,57000,49000,47000,45000]
years = [2015,2016,2017,2018,2019,2020]
plt.plot(years , price)
```

```python
# from a pandas datframe
batsman = pd.read_csv('sharma-kohli.csv')
batsman
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# mutiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
# labels title
plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Ipl Seasons')
plt.ylabel('Runs Scored')
```

```python
#  colors (hex) and line ( width and height )
# mutiple plots
plt.plot(batsman['index'],batsman['V Kohli'] ,color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#ef6b1d')
# labels title
plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Ipl Seasons')
plt.ylabel('Runs Scored')
```

```python
# line style
#  colors (hex) and line ( width and height )
# mutiple plots
# plt.plot(batsman['index'],batsman['V Kohli'] ,color='green',linestyle='dashed')
# plt.plot(batsman['index'],batsman['V Kohli'] ,color='green',linestyle='dashdot')
plt.plot(batsman['index'],batsman['V Kohli'] ,color='green',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#ef6b1d',linestyle='dotted',linewidth=2,marker='+')
# labels title
plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Ipl Seasons')
plt.ylabel('Runs Scored')
```

```python
# legend -> location
plt.plot(batsman['index'],batsman['V Kohli'] ,color='green',label='VIRAT')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#ef6b1d',label='ROHIT')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Ipl Seasons')
plt.ylabel('Runs Scored')

plt.legend(loc='upper right')
```

```python
# limiting axes
price = [48000,54000,57000,49000,47000,45000,4500000]
years = [2015,2016,2017,2018,2019,2020,2021]
plt.plot(years , price)
# graph has outlier

plt.ylim(0,175000)
plt.xlim(2017,2019)
```

```python
# grid

plt.plot(batsman['index'],batsman['V Kohli'] ,color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#ef6b1d')
# labels title
plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Ipl Seasons')
plt.ylabel('Runs Scored')

plt.grid()
```

```python
# show
plt.plot(batsman['index'],batsman['V Kohli'] ,color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#ef6b1d')
# labels title
plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Ipl Seasons')
plt.ylabel('Runs Scored')

plt.show()
```

**BAR CHART**

```python
# simple bar chart
children = [30,50,20,10,5]
colors = ['blue','green','red','yellow','pink']
# plt.bar(colors,children,color='black')
plt.bar(colors,children,color=['blue','green','red','yellow','pink'])
```

```python
# horizontal bar chart
plt.barh(colors,children,color=['blue','green','red','yellow','pink'])
```

```python
# color and label
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'],df['2015'])
```

```python
# multiple bars side by side
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2)

plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

```python
children = [10,20,40,10,30]
colors = ['red red red red','green green green green','blue blue blue blue','yellow yellow yellow yellow','pink pink pink pink']
plt.bar(colors,children)
plt.xticks(rotation='vertical')
plt.xticks(rotation=70)
```

```python
# stacked bar chart
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom = df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom = df['2017'] + df['2016'],label='2015')
plt.legend()
plt.show()
```

```python
df = pd.read_csv('gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
```
