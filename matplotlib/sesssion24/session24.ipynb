```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# colored scatterplot
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0 , 'Iris-versicolor':1 , 'Iris-virginica':2})
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=0.9)
plt.xlabel('sepal length')
plt.ylabel('petal length')
plt.colorbar()
```

```python
# plot size
plt.figure(figsize=(15,7))
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=1)
plt.xlabel('sepal length')
plt.ylabel('petal length')
plt.colorbar()
```

```python
# annotations
batters = pd.read_csv('batter.csv')
batters.head()
```

```python
sample_df = batters.head(100).sample(25,random_state=5)
sample_df
```

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
for i in range (sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
plt.scatter(x,y)
plt.text(1,5,'Point 1')
plt.text(2,6,'Point 2')
plt.text(3,7,'Point 3')
plt.text(4,8,'Point 4',fontdict={"size" : 12 , "color":"blue"})
```

```python
# horizontal and vertical lines
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
plt.axhline(130,color='red')
plt.axhline(140,color='green')
plt.axvline(30,color='red')
for i in range (sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
# subplots
batters.head()
plt.scatter(batters['avg'],batters['strike_rate'])
```

```python
fig , ax = plt.subplots(figsize=(15,6))
ax.scatter(batters['avg'],batters['strike_rate'])
ax.set_title('something')
ax.set_xlabel('avg')
ax.set_ylabel('strike rate ')
```

```python
fig,ax=plt.subplots(nrows=2,ncols=1,sharex=True)
ax[0].scatter(batters['avg'],batters['strike_rate'])
ax[1].scatter(batters['avg'],batters['runs'])
ax[0].set_title('avg vs strike rate')
ax[1].set_xlabel('avg')
ax[0].set_ylabel('strike rate')
ax[1].set_ylabel('runs')
ax[1].set_title('avg vs runs')
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2)
ax[0,0].scatter(batters['avg'],batters['strike_rate'])
ax[0,1].scatter(batters['avg'],batters['runs'])
ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['runs'])
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1) # 2 row 1 col 1st graph
ax1.scatter(batters['avg'],batters['runs'])
ax2 = fig.add_subplot(212) # 2 row 1 col 2nd graph
ax2.scatter(batters['avg'],batters['strike_rate'])
```

```python
# 3d scatter plots
fig = plt.figure()
ax = plt.subplot(projection='3d')
ax.scatter3D(batters['runs'],batters['avg'],batters['strike_rate'],marker='+')
ax.set_title('Ipl batsman analysis')
ax.set_xlabel('runs')
ax.set_ylabel('avg')
ax.set_zlabel('strike rate')
```

```python
# 3d line plot
x = [0,1,5,25]
y = [0,10,13,0]
z = [0,13,20,9]
fig = plt.figure()
fig = plt.figure()
ax = plt.subplot(projection='3d')
ax.scatter3D(x,y,z,s=[100,100,100,100])
ax.plot3D(x,y,z,color='green')
```

```python
# 3d surface plots
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx , yy = np.meshgrid(x,y)
xx.shape
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.log(yy)
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
# contour graph
# 3d graph to 2d
z = xx**2 + yy**2
z.shape
fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = xx**2 + yy**2
z.shape
fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```
