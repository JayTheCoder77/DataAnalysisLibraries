```python
# Merge - imp interview question
# inner join - find common items which are present in both tables
students
regs
students.merge(regs,how='inner',on='student_id')
```

```python
# left join - common items and also the items in left table but not in right table
courses.merge(regs,how='left',on='course_id')
```

```python
# right join - common items and also the items in right table but not in left table
temp_df = pd.DataFrame({
    'student_id' : [26,27,28],
    'name' : ['Jay','Skillf','Reaver'],
    'partner' : [28,26,17]
})
students = pd.concat([students,temp_df],ignore_index=True)
students.merge(regs,how='right',on='student_id')
```

```python
students.merge(regs,how='left',on='student_id')
```

```python
# outer join - all the values are included
students.merge(regs,how='outer',on='student_id')
```
