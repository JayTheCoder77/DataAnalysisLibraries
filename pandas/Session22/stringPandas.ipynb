```python
# problem in vanilla python
s = ['cat','mat',None,'rat']
[i.startswith('c') for i in s] # fails as None is present also list comprehension is slow
```

```python
# how does pandas solve this issue?
s = pd.Series(['cat','mat',None,'rat'])
# str = string accessor
s.str.startswith('c')
# fast and optimized and handles special cases
```

```python
# import titanic
df = pd.read_csv('titanic.csv')
name = df['Name']
```

```python
# common functions
# lower / upper / capitalize
name.str.lower()
name.str.upper()
name.str.capitalize()
name.str.title()

 # len
name.str.len().max()
name[name.str.len() == 82].values[0]
```

```python
# strip
"                           jay                     ".strip()
```

```python
name.str.strip()
```

```python
# split -> get
df['lastname'] = name.str.split(',').str.get(0)
```

```python
# name.str.split(',').str.get(1).str.strip().str.split(' ',n=1).str.get(0)
# name.str.split(',').str.get(1).str.strip().str.split(' ',n=1).str.get(1)
df[['title','first_name']] = name.str.split(',').str.get(1).str.strip().str.split(' ',n=1,expand=True)
df
df['title'].value_counts()
```

```python
# replace
df['title'] = df['title'].str.replace('Ms.','Miss.')
df['title'] = df['title'].str.replace('Mlle.','Miss.')
```

```python
df['title'].value_counts()
```

```python
# filtering
df[df['first_name'].str.startswith('A')]
df[df['first_name'].str.endswith('A')]
```

```python
# isdigit / isalpha
df[df['first_name'].str.isdigit()]
df[df['first_name'].str.isalpha()]
```

```python
# applying ranges
# contains
df[df['first_name'].str.contains('john',case=False)].head(3)
```

```python
# find lastnames with start and end char - vowel
df[df['lastname'].str.contains('[aeiouAEIOU].+[aeiouAEIOU]$')]
```

```python
# slicing
df['Name'].str[:4]
df['Name'].str[::2]
df['Name'].str[::-1] # reverse
```
