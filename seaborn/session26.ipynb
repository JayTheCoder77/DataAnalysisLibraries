```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

## Categorical Plots

### Categorical Scatter Plot

- Stripplot
- Swarmplot

### Categorical Distribution Plots

- Boxplot
- Violinplot

### Categorical Estimate Plot -> for central tendency

- Barplot
- Pointplot
- Countplot

### Figure level function -> `catplot`

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip')
```

```python
# strip plot
sns.stripplot(data=tips,x='day',y='total_bill',jitter=False)
```

```python
# using catplot figure lvl function
sns.catplot(data=tips,x='day',y='total_bill',kind='strip')
```

```python
# jitter
sns.catplot(kind='strip',data=tips,x='day',y='total_bill',jitter=0.4,hue='sex')
```

```python
# swarmplot
sns.catplot(kind='swarm',data=tips,x='day',y='total_bill')
```

```python
# box plot
sns.boxplot(data = tips , x = 'day' , y = 'total_bill')
```

```python
sns.catplot(hue='sex',data=tips,x='day',y='total_bill',kind='box')
```

```python
sns.boxplot(data=tips,y='total_bill')
```

```python
# violin plot = boxplot + kdeplot
sns.violinplot(data=tips,x='day',y='total_bill')
```

```python
sns.catplot(kind='violin',hue='sex',data=tips,x='day',y='total_bill',split=True)
```

```python
# barplot
sns.barplot(data=tips,x='sex',y='total_bill')
```

```python
import numpy as np
sns.catplot(kind='bar',data=tips,x='sex',y='total_bill',errorbar=None,estimator=np.std)
sns.catplot(kind='bar',data=tips,x='sex',y='total_bill',errorbar=('ci',95),hue='smoker')
```

```python
# pointplot
sns.pointplot(data=tips,x='sex',y='total_bill',hue='smoker')
sns.catplot(kind='point',data=tips,x='sex',y='total_bill',hue='smoker',errorbar=None)
```

```python
sns.catplot(kind='point',data=tips,x='sex',y='total_bill')
```

```python
# countplot
sns.countplot(data=tips,x='sex',hue='day')
```

```python
# faceting using catplot
sns.catplot(data=tips,x='sex',y='total_bill',col='smoker',kind='box',row='time')
```

```python
# regression plots
```

* **reg plot**
* **lm plot**

```python
sns.scatterplot(data=tips,x='total_bill',y='tip')
```

```python
# axes lvl function
sns.regplot(data=tips,x='total_bill',y='tip') # hue param not available here
```

```python
# lmplot -> figure lvl function
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex')
```

```python
# resid plot - residual plot

sns.residplot(data=tips,x='total_bill',y='tip')
```

```python
# facet grid
# fair grid
# joint grid
```

```python
sns.catplot(data=tips,x='sex',y='total_bill',kind='violin',col='day',row='time')
```

```python
g = sns.FacetGrid(data=tips,col='day',row='time',hue='smoker')
g.map(sns.scatterplot,'sex','total_bill')
g.add_legend()
```

```python
# pair plot
sns.pairplot(iris,hue='species')
sns.pairplot(iris)
```

```python
# pair grid
g = sns.PairGrid(data=iris,hue='species')
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.boxplot)
g.map_offdiag(sns.histplot)
```

```python
g = sns.PairGrid(data=iris,hue='species',vars=['sepal_length','petal_length'])
g.map_diag(sns.boxplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.histplot)
```

```python
# joint grid and joint plot
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',kind='kde' , hue='sex')
sns.jointplot(data=tips,x='total_bill',y='tip',kind='hist')
sns.jointplot(data=tips,x='total_bill',y='tip',kind='reg')
```

```python
g = sns.JointGrid(data=tips,x='total_bill',y='tip')
g.plot(sns.scatterplot,sns.kdeplot)
```
