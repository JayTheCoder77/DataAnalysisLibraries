```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
#scatter plot  -> axes level function
sns.scatterplot(data = tips,x='total_bill',y='tip',style='time')
```

```python
# relplot - > figure level function
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='sex',style='time',size='size')
```

```python
# line plot
gap = px.data.gapminder()
temp_df = gap[gap['country'] == 'India']
```

```python
sns.lineplot(data=temp_df,x='year',y='lifeExp')
```

```python
# relplot
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line')
```

```python
# hue -> style
temp_df = gap[gap['country'].isin(['India','France','China'])]
temp_df
```

```python
sns.relplot(kind='line',data=temp_df,x='year',y='lifeExp',hue='country',style='continent',size='continent')
```

```python
# facet plot -> figure level function works only with relplot not scatterplot lineplot etc
tips
# sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',row='sex')
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='smoker',row='sex')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='sex',row='day')
```

```python
# col wrap
sns.relplot(data=gap,x='lifeExp',y='gdpPercap',kind='scatter',col='year',col_wrap=4)
```

```python
# distributional plots
```

```python
# figure lvl function - > displot
# axes lvl  ->  histplot , kdeplot , rugplot
```

```python
tips
sns.histplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,kind='hist',x='total_bill')
```

```python
sns.displot(data=tips,kind='hist',x='total_bill',bins=20)
```

```python
# categorical cols plot
# this is kind of a countplot
sns.displot(data=tips,kind='hist',x='day')
```

```python
# hue parameter
sns.displot(data=tips,kind='hist',x='tip',hue='sex')
```

```python
# element -> step
sns.displot(data=tips,kind='hist',x='tip',hue='sex',element='step')
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.displot(data=titanic,kind='hist',x='age',element='step',hue='sex')
```

```python
# faceting usinf row col
sns.displot(data=titanic,kind='hist',x='age',element='step',col='sex')
```

```python
# kde plot -> kernel density estimate
# uses gausian kernel -> also estimates/predicts missing values
sns.kdeplot(data=tips,x='total_bill')
```

```python
# hue -> fill
sns.displot(data=tips,kind='kde',x='total_bill',hue='sex',fill=True)
```

```python
# rugplot
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

```python
# bivariate histogram -> two cols
# sns.histplot(data=tips,x='total_bill',y='tip')
sns.displot(data=tips,kind='hist',x='total_bill',y='tip')
```

```python
# bivariate kdeplot
# sns.kdeplot(data=tips,x='total_bill',y='tip')
sns.displot(data=tips,kind='kde',x='total_bill',y='tip')
```

```python
# matrix plot
# heatmap and clustermap
```

```python
temp_df = gap.pivot(index='country',columns='year',values='lifeExp')
# heatmap is a axes lvl function
plt.figure(figsize=(15,15))
sns.heatmap(data=temp_df)
```

```python
temp_df = gap[gap['continent'] == 'Europe'].pivot(index='country',columns='year',values='lifeExp')
plt.figure(figsize=(15,15))
sns.heatmap(data=temp_df,annot=True,linewidth=0.5,cmap='summer')
```

```python
# cluster map
iris= px.data.iris()
iris
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```

```python
# change size in fig lvl
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True,height=10,aspect=2)
```
