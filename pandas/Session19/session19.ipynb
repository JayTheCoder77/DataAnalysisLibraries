```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('imdb-top-1000.csv')
movies.head()
```

```python
genres = movies.groupby('Genre')
genres
```

```python
# Applying builtin aggregation fuctions on groupby objects
genres.sum(numeric_only=True)
genres.min(numeric_only=True)
genres.mean(numeric_only=True)
genres.std(numeric_only=True)
```

```python
# find the top 3 genres by total earning
genres.sum()['Gross'].sort_values(ascending=False).head(3)
```

```python
movies.groupby('Genre')['Gross'].sum().sort_values(ascending=False).head(3)
# this is more correct as we filter group first then sum
```

```python
# genre with highest avg imdb rating
movies.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head(1)
```

```python
# director with highest popularity
# based on no of votes
movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head(1)
```

```python
# highest rated movie of each genre
# movies.groupby('Genre')['IMDB_Rating'].max().sort_values(ascending=False)
movies[movies.IMDB_Rating == movies.groupby('Genre')['IMDB_Rating'].transform('max')].sort_values(by='IMDB_Rating', ascending=False)[['Genre', 'IMDB_Rating', 'Series_Title']]
```

```python
# agg method
# passing dict
genres.agg(
    {
        'Runtime' : 'mean',
        'IMDB_Rating' : 'mean',
        'No_of_Votes' : 'sum',
        'Gross' : 'sum',
        'Metascore' : 'min'
    }
)
```

```python
# agg -> passing list
numeric = genres.obj.select_dtypes(include=np.number).columns
genres[numeric].agg(['min', 'max', 'mean'])
```

```python
# merge list and dict in agg
genres.agg(
    {
        'Runtime' : ['mean','min','max'],
        'IMDB_Rating' : 'mean',
        'No_of_Votes' : 'sum',
        'Gross' : 'sum',
        'Metascore' : 'min'
    }
)
```

```python
# looping on groups
genres
for group , data in genres:
    # print(type(group) , type(data))
    print(group , data)
```

```python
# highest rated movie of each genre
df = pd.DataFrame(columns = movies.columns)
for group , data in genres :
  # deprecated -> df = df.append(data[data['IMDB_Rating'] == data['IMDB_Rating'].max()])
  df = pd.concat([df , data[data['IMDB_Rating'] == data['IMDB_Rating'].max()]]) # Pass both DataFrames in a list
df
```

```python
# split (Apply) combine
# apply builtin function
genres.apply(min)
```

```python
# find movies starting with a
def moviesA(group):
  return group['Series_Title'].str.startswith('A').sum()
genres.apply(moviesA)
```

```python
# find rank of movie according to group imdb
def rank(group):
  group['genre_rank'] = group['IMDB_Rating'].rank(ascending=False)
  return group
genres.apply(rank)
```

```python
# find normalized imdb rating group wise
def normalize(group):
  group['normalized_rank'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min()) / (group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
  return group
genres.apply(normalize)
```

```python
# groupby on multiple cols
duo = movies.groupby(['Director','Star1'])
duo
# size
duo.size()
# get_group
duo.get_group(('Christopher Nolan','Christian Bale'))
```

```python
# most earning actor-> director combo
duo['Gross'].sum().sort_values(ascending=False)
```

```python
# actor -> genre combo for metascore
movies.groupby(['Star1','Genre'])['Metascore'].mean().reset_index().sort_values('Metascore', ascending=False)
```

```python
# agg on multiple groupby
numeric = duo.obj.select_dtypes(include=np.number).columns
duo[numeric].agg(['min', 'max', 'mean'])
```

Excercise ON IPL dataset

```python
ipl = pd.read_csv('deliveries.csv')
ipl
```

```python
ipl.shape
```

```python
# top 10 batsman with most runs
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# batsman with most num of sixes
six = ipl[ipl['batsman_runs'] == 6]
# below two give same result
six.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(10)
six.groupby('batsman')['batsman'].value_counts().sort_values(ascending=False).head(10)
```

```python
# batsman with most sixes in last 5 overs
over = ipl[ipl['over'] > 15]
sixOrFour = over[(over['batsman_runs'] == 6) | (over['batsman_runs'] == 4)]
sixOrFour.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(10)
```

```python
# v kohli record against all teams
vk = ipl[ipl['batsman'] == 'V Kohli']
vk.groupby('bowling_team')['batsman_runs'].sum().reset_index()
```

```python
# create function that returns highest score of batsman
def highestScore(batsman):
  batter = ipl[ipl['batsman'] == batsman]
  return batter.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1)
highestScore('RG Sharma')
highestScore('V Kohli')
highestScore('MS Dhoni')
```
