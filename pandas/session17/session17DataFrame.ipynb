Creating DataFrames

```python
# lists
student_data = [
    [100,80,10],
    [90,70,7],
    [80,50,2],
    [120,100,14]
]

pd.DataFrame(student_data,columns=['iq','marks','package'])
```

```python
# DICT
student_dict = {
    'iq' : [100,90,80,120,0,0],
    'marks' : [80,70,100,50,0,0],
    'package': [10,7,15,2,0,0]
}
students = pd.DataFrame(student_dict)
students
```
