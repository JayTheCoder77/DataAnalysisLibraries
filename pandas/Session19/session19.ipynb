```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('imdb-top-1000.csv')
movies.head()
```

```python
genres = movies.groupby('Genre')
genres
```

```python
# Applying builtin aggregation fuctions on groupby objects
genres.sum(numeric_only=True)
genres.min(numeric_only=True)
genres.mean(numeric_only=True)
genres.std(numeric_only=True)
```

```python
# find the top 3 genres by total earning
genres.sum()['Gross'].sort_values(ascending=False).head(3)
```

```python
movies.groupby('Genre')['Gross'].sum().sort_values(ascending=False).head(3)
# this is more correct as we filter group first then sum
```

```python
# genre with highest avg imdb rating
movies.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head(1)
```

```python
# director with highest popularity
# based on no of votes
movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head(1)
```

```python
# highest rated movie of each genre
# movies.groupby('Genre')['IMDB_Rating'].max().sort_values(ascending=False)
movies[movies.IMDB_Rating == movies.groupby('Genre')['IMDB_Rating'].transform('max')].sort_values(by='IMDB_Rating', ascending=False)[['Genre', 'IMDB_Rating', 'Series_Title']]
```

```python
# no of movies done by each actor
movies['Star1'].value_counts()
```

```python
# using group by
movies.groupby('Star1')['Series_Title'].count().sort_values(ascending=False)
```

```python
# agg method
# passing dict
genres.agg(
    {
        'Runtime' : 'mean',
        'IMDB_Rating' : 'mean',
        'No_of_Votes' : 'sum',
        'Gross' : 'sum',
        'Metascore' : 'min'
    }
)
```

```python
# agg -> passing list
numeric = genres.obj.select_dtypes(include=np.number).columns
genres[numeric].agg(['min', 'max', 'mean'])
```

```python
# merge list and dict in agg
genres.agg(
    {
        'Runtime' : ['mean','min','max'],
        'IMDB_Rating' : 'mean',
        'No_of_Votes' : 'sum',
        'Gross' : 'sum',
        'Metascore' : 'min'
    }
)
```

```python
# looping on groups
genres
for group , data in genres:
    # print(type(group) , type(data))
    print(group , data)
```

```python
# highest rated movie of each genre
df = pd.DataFrame(columns = movies.columns)
for group , data in genres :
  # deprecated -> df = df.append(data[data['IMDB_Rating'] == data['IMDB_Rating'].max()])
  df = pd.concat([df , data[data['IMDB_Rating'] == data['IMDB_Rating'].max()]]) # Pass both DataFrames in a list
df
```
